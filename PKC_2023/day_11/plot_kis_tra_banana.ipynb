* Assignment: How to change the title of x and y axis..

```python
g = sns.FacetGrid(kashti, hue="survived", col="pclass" , margin_titles=True,
                  palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter , "fare" , "age" , edgecolor="w").add_legend()
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="sex" , margin_titles=True,
                  palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter , "fare" , "age" , edgecolor="w").add_legend()
```

```python
sns.stripplot(data=kashti , x="sex" , y="fare" , hue="class")
```

```python
# HEat Map
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("HEat map ka plot")
```

* Positive Correlation mean the value is between 0 to 1
* Negative Correlation mean the value is between -1 to 0
