# Ab yahain pr hum age group banaana ja ra hun 
## Like 0-10 sa baby 10-20 sa adults etc..

### Grouping

## Sab sa pehla ap na apna columns dekh lena ha

```python
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'] , [0,10,18,30,50,80] , 
    labels=['kids','child','Youth','Middle_age','Old_age'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind='bar')
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived', vars=['age','fare','pclass']) # vars mean ka in in varaibles ka pairplot banana ha
```
